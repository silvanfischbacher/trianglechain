```python
from trianglechain import TriangleChain
import numpy as np
%load_ext autoreload
%autoreload 2
```

```python
samples = np.load('chain.npy')
lnprobs = np.load('lnprobs.npy')
labels = [r'$\Omega_m$', r'$S_8$', r'$A_1$']#r'$n$', r'$h$', r'$\Omega_b$', r'$A_1$',
          #r'$\delta_{z,1}$',r'$\delta_{z,2}$',r'$\delta_{z,3}$',r'$\delta_{z,4}$',r'$\delta_{z,5}$']
```

# Credible Intervals

```python
#default 0.68
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True);
```

```python
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True, credible_interval=0.95);
```

# Levels

```python
#default highest density interval hdi
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True);
```

```python
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True, levels_method = 'percentile');
```

```python
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True, levels_method = 'PJ_HPD', lnprobs=lnprobs); #requires lnprobs
```

# bestfit

```python
#default mode
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True);
```

```python
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True, bestfit_method='mean');
```

```python
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True, bestfit_method='median');
```

```python
tri = TriangleChain(labels=labels, params=['omega_m', 'S8', 'A_IA'])
tri.contour_cl(samples, show_values=True, bestfit_method='best_sample', lnprobs=lnprobs);#requires lnprobs
```
